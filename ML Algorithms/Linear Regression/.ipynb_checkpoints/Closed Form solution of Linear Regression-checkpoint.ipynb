## Closed Form Solution of Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

```python
#create dataset
X,Y=make_regression(n_samples=400,n_features=1,n_informative=1,noise=3,random_state=11)
print(X.shape,Y.shape)
Y=Y.reshape((-1,1))
print(Y.shape)
```

```python
#normalise
X=(X-X.mean())/X.std()
plt.figure()
plt.scatter(X,Y)
plt.style.use("seaborn")
plt.title("Normalised Data")
plt.show()
```

```python
#add ones to X matrix
ones=np.ones((X.shape[0],1))
X_=np.hstack((X,ones))
print(X_.shape)
print(X_[:5])
```

```python
#prediction
def predict(X,theta):
    return np.dot(X,theta)

#closed form solution implementation
def getThetaClosedForm(X,Y):
    Y=np.mat(Y)
    firstPart=np.dot(X.T,X)
    secondPart=np.dot(X.T,Y)
    theta=np.linalg.pinv(firstPart)*secondPart
    return theta
```

```python
theta=getThetaClosedForm(X_,Y)
print(theta)
```

```python
#plot the predictions
plt.scatter(X,Y)
plt.plot(X,predict(X_,theta),color='red',label='prediction')
plt.legend()
plt.title("Closed form solution")
plt.show()
```
