```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../Datasets/MNIST-2/mnist_train.csv')
print(df.shape)
print(df.columns)
print(df.head(n=5))
```

```python
data=df.values
print(data.shape)
print(type(data))
```

```python
X=data[:,1:]
Y=data[:,0]
print(X.shape,Y.shape)
```

```python
split=int(0.8*X.shape[0])
print(split)
X_train=X[:split]
Y_train=Y[:split]
X_test=X[split:]
Y_test=Y[split:]
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```

```python
def drawImg(sample):
    img=sample.reshape((28,28))
    plt.imshow(img,cmap='gray')
    plt.show()

drawImg(X_train[3])    
print(Y_train[3])
```

```python
#knn algo
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))

def knn(X,Y,queryPoint,k=5):
    vals=[]
    m=X.shape[0]
    for i in range(m):
        d=dist(queryPoint,X[i])
        vals.append((d,Y[i]))
    vals=sorted(vals)
    vals=vals[:k]
    vals=np.array(vals)
    new_vals=np.unique(vals[:,1],return_counts=True)
    index=new_vals[1].argmax()
    pred=new_vals[0][index]
    return pred
```

```python
pred=knn(X_train,Y_train,X_test[0])
print(int(pred))
drawImg(X_test[0])
print(Y_test[0])
```

```python
def check_accuracy():
    m=X_test.shape[0]
    Y_pred=[]
    for i in range(m):
        print(i)
        p=knn(X_train,Y_train,X_test[i])
        Y_pred.append(p)
    Y_pred=np.array(Y_pred)
    return ((Y_test==Y_pred).sum())/m
```

```python
ans=check_accuracy()
print(ans)
```
