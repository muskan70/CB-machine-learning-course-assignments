```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfx=pd.read_csv('../Datasets/Locally Weighted Regression/weightedX.csv')
dfy=pd.read_csv('../Datasets/Locally Weighted Regression/weightedY.csv')
X=dfx.values
Y=dfy.values
print(X.shape,Y.shape)
```

```python
#Normalize the data
X=(X-X.mean())/X.std()
plt.style.use("seaborn")
plt.title("Normalized Data")
plt.scatter(X,Y)
plt.show()
```

```python
#generate W(m*m) --- diagonal matrix
def getW(query_point,X,tau):
    M=X.shape[0]
    W=np.mat(np.eye(M))
    for i in range(M):
        xi=X[i]
        x=query_point
        W[i,i]=np.exp(np.dot((xi-x),(xi-x).T)/(-2*tau*tau))
    return W    
```

```python
X=np.mat(X)
Y=np.mat(Y)
W=getW(-1,X,1)
M=X.shape[0]
print(W.shape)
print(W)
```

```python
def predict(X,Y,query_x,tau):
    ones=np.ones((M,1))
    X_=np.hstack((X,ones))
    qx=np.mat((query_x,1))
    W=getW(qx,X_,tau)
    
    theta=np.linalg.pinv(X_.T*(W*X_))*(X_.T*(W*Y))
    pred=np.dot(qx,theta)
    return theta,pred
```

```python
theta,pred=predict(X,Y,1.0,1.0)
print(theta)
print(pred)
```

```python
#visualize
def plotPrediction(tau):
    X_test=np.linspace(-2,2,20)
    Y_test=[]
    for xq in X_test:
        theta,pred=predict(X,Y,xq,tau)
        Y_test.append(pred[0][0])
    Y_test=np.array(Y_test)
    X0=np.array(X)
    Y0=np.array(Y)
    plt.scatter(X0,Y0)
    plt.scatter(X_test,Y_test,color='red')
    plt.show()
```

```python
taus=[0.1,0.5,1,5,10]
for i in taus:
    plotPrediction(i)
```
