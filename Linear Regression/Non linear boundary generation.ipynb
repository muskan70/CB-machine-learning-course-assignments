# Non-linear boundary using linear model

```python
#importing dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dfx=pd.read_csv('../Datasets/Locally Weighted Regression/weightedX.csv')
dfy=pd.read_csv('../Datasets/Locally Weighted Regression/weightedY.csv')
x=dfx.values
y=dfy.values
print(x.shape,y.shape)
```

```python
x=x.reshape((-1,))
y=y.reshape((-1,))
print(x.shape,y.shape)
```

```python
#Increase model complexity by adding more features
Y=y
x1=x
x2=x**2
X=np.stack((x1,x2),axis=1)
print(X.shape,Y.shape)
```

```python
#plot data
plt.scatter(X[:,0],Y)
plt.show()
```

```python
#apply linear regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X,Y)
output=model.predict(X)
bias=model.intercept_
coef=model.coef_
print(bias)
print(coef)
model.score(X,Y)
```

```python
#plot predicted result
plt.scatter(X[:,0],Y,label='data')
plt.scatter(X[:,0],output,color='orange',label='prediction')
plt.legend()
plt.show()
```
