# Gradient descent Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd

dfx=pd.read_csv('../Datasets/Linear Regression/linearX.csv')
dfy=pd.read_csv('../Datasets/Linear Regression/linearY.csv')
x=dfx.values
y=dfy.values

X=(x-x.mean())/x.std()
Y=y

#Create a meshgrid
T0=np.arange(-2,3,0.01)
T1=np.arange(-2,3,0.01)
T0,T1=np.meshgrid(T0,T1)
J=np.zeros(T0.shape)
m=T0.shape[0]
n=T0.shape[1]
for i in range(m):
    for j in range(n):
        J[i,j]=np.sum((Y-T1[i,j]*X-T0[i,j])**2)   #ERROR calculation at each value of theta-(theta0,theta1)
print(J.shape)        
```

```python
#plot 3D Error:Loss
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.plot_surface(T0,T1,J,cmap='rainbow')
plt.show()
```

```python
#plot 3D contours
fig=plt.figure()
axes=fig.gca(projection='3d')
axes.contour(T0,T1,J,cmap='coolwarm')
plt.show()
```
