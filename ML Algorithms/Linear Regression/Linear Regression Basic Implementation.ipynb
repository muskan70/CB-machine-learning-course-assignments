# Linear Regression Implementation

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#get the training data
dfx=pd.read_csv('../../Datasets/Linear Regression/linearX.csv')
dfy=pd.read_csv('../../Datasets/Linear Regression/linearY.csv')
x=dfx.values
y=dfy.values
print(x.shape)
print(y.shape)
```

```python
#preprocess and visualize
X=(x-x.mean())/x.std()
Y=y
plt.scatter(X,Y)
plt.show()
```

```python
#generate the hypothesis:line
def hypothesis(x,theta):
    return theta[0]+theta[1]*x

#generate error
def error(X,Y,theta):
    m=X.shape[0]
    err=0
    for i in range(m):
        hx=hypothesis(X[i],theta)
        err+=(hx-y[i])**2
    return err

#generate gradient
def gradient(X,Y,theta):
    m=X.shape[0]
    grad=np.zeros((2,))
    for i in range(m):
        hx=hypothesis(X[i],theta)
        grad[0]+=(hx-Y[i])
        grad[1]+=(hx-Y[i])*X[i]
    return grad    
```

```python
#algo--Gradient Descent
def gradientDescent(X,Y,learning_rate=0.001,max_itr=100):
    theta=np.array([2.0,3.0])
    itr=0
    error_list=[]
    theta_list=[]
    
    while(itr<=max_itr):
        grad=gradient(X,Y,theta)
        err=error(X,Y,theta)
        error_list.append(err)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        theta_list.append((theta[0],theta[1]))
        itr+=1
    return theta,error_list,theta_list 
```

```python
def linearRegressionPredictor(X,Y,learning_rate,max_itr):
    final_theta,error_list,theta_list=gradientDescent(X,Y,learning_rate,max_itr)

    #plotting the final result
    plt.plot(error_list)
    plt.show()
    print(final_theta)
    
    #get the test data
    x_test=np.linspace(-2,6,10)
    print(x_test)
    
    #plot line for test data
    plt.scatter(X,Y,label='Training Data')
    plt.plot(x_test,hypothesis(x_test,final_theta),color='orange',label='Prediction')
    plt.legend()
    plt.show()
```

```python
#max_itr=100 , learning_rate=0.001
linearRegressionPredictor(X,Y,0.001,100)
```

```python
#changing max_itr=50
linearRegressionPredictor(X,Y,0.001,50)
```

```python
#changing max_itr=10
linearRegressionPredictor(X,Y,0.001,10)
```

```python
#changing learning rate=0.01
linearRegressionPredictor(X,Y,0.01,100)
```

```python
#changing learning rate=0.1
linearRegressionPredictor(X,Y,0.1,100)
```
